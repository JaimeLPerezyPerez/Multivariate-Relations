```python
import pandas as pd
from sdv.datasets.local import load_csvs
faults = [0,2,6,8,17,20,23,24,25,26,27,28,31,32,33,34,35,38,39,40,41,43,44,45,48,49]
goods = []
for i in range(51):
    if(i not in faults):
        goods.append(i)
```

```python

datasets = load_csvs(folder_name='./data')
thingies = [pd.DataFrame() for i in range(51)]
time = datasets["Sensors-Thingy-color_b-sampled10min"]["time_ms"]
for x in datasets:
    datasets[x] = datasets[x].drop("time_ms",axis = 1)
    y = 0
    for i in datasets[x]:
        thingies[y][x] = datasets[x][i]
        y = y + 1

columns = ["color b", "color c", "color g", "color r", "eCO2","humidity", "pressure", "rssi", "sound", "temperature", "TVOC", "voltage"]
for x in thingies:
    x.columns = columns

new_dataframe = pd.DataFrame()
for x in range(51):
    if(x not in faults):
        thingies[x]["id"] = [x for i in range(len(thingies[x]))]
        thingies[x]["time"] =time
        new_dataframe = pd.concat([new_dataframe,thingies[x]])
    

new_dataframe
```

```python
new_dataframe.to_csv("./data_full.csv",index= False)
```
