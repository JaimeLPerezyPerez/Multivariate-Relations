```python
day = False
if(day):
    #1 day
    timeFrame = 144
else:
    #5hours
    #timeFrame = 30 
    #1 week
    timeFrame = 1008
```

```python
i = 0
```

```python
i = i - timeFrame
```

```python
#plot select columns over timeframe
light_CO2 = ["TVOC","eCO2","color b", "color c", "color g", "color r"]
temp_CO2 = ["eCO2","temperature"]
temp_volt = ["voltage","temperature"]
tvoc_CO2 = ["TVOC","eCO2"]
light = ["color b", "color c", "color g", "color r"]
humidity = ["eCO2","temperature","humidity"]
light_voltage = ["voltage","color b", "color c", "color g", "color r"]
columns_to_plot = light
data = pd.read_csv("./PAR10000.csv")
SENSOR = 0
new_data = data[data["id"] == SENSOR]
fig, axs = plt.subplots(len(columns_to_plot),1, figsize = (16,9))
axis = 0
for x in columns_to_plot:
    if (x != "id" and x != "time"):
        axs[axis].plot(new_data[x][i:i + timeFrame].tolist())
        axs[axis].set_title(x)
        axis = axis + 1
    
    

fig.tight_layout()
plt.show()
i = i + timeFrame
```

```python
#plot all columns over timeframe
data = pd.read_csv("./PAR_all.csv")
SENSOR = 5
new_data = data[data["id"] == SENSOR]
fig, axs = plt.subplots(len(new_data.columns) - 2,1,figsize = (50,50))
axis = 0
for x in new_data:
    if (x != "id" and x != "time"):
        axs[axis].plot(new_data[x][i:i + timeFrame])
        axs[axis].set_title(x)
        axis = axis + 1
    

fig.tight_layout()
plt.show()
i = i + timeFrame
```
